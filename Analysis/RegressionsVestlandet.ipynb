```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor



#Catchment parameters for Vestlandet stations:

params = vf_params.copy()
params["stID"] = params["stID"].astype(str).str.strip()

params_vest = params[params["stID"].isin(vestlandet_list)].copy()
params_vest["region"] = "Vestlandet"

#Merging parameters with log AM7
df_vest = params_vest.merge(am7_vest_df, on="stID", how="inner")




# Dataframe for VIF + OLS
exclude = {"stID", "region", "log_AM7"}
num_cols = df_vest.select_dtypes(include=["number"]).columns
predictor_cols = [c for c in num_cols if c not in exclude]

# drop rows with missing predictors/targets 
df_ols = df_vest.dropna(subset=predictor_cols + ["log_AM7"]).copy()


#Checking the VIF for the predictors  

def compute_vif(df, predictors):
    X = sm.add_constant(df[predictors].astype(float))
    vals = []
    for i, var in enumerate(predictors, start=1):  
        vals.append((var, variance_inflation_factor(X.values, i)))
    return pd.DataFrame(vals, columns=["variable", "VIF"]).sort_values("VIF", ascending=False)

def vif_filter(df, predictors, thresh=10.0):
    vars_left = predictors.copy()
    while True:
        vif_df = compute_vif(df, vars_left)
        max_vif = vif_df["VIF"].max()
        worst_var = vif_df.loc[vif_df["VIF"].idxmax(), "variable"]
        print(f"Max VIF = {max_vif:.2f} ({worst_var})")
        if max_vif > thresh:
            print(f"  -> dropping {worst_var}")
            vars_left.remove(worst_var)
        else:
            print("All VIFs below threshold.")
            break
    final_vif = compute_vif(df, vars_left)
    return vars_left, final_vif


predictors_vif, vif_table = vif_filter(df_ols, predictor_cols, thresh=10.0)

print("\nPredictors after VIF filtering:", predictors_vif)
print("\nFinal VIF table:")
print(vif_table)

#Running the OLS regression model with VIF filtered predictors 
X_final = sm.add_constant(df_ols[predictors_vif].astype(float))
y_final = df_ols["log_AM7"]

ols_vest = sm.OLS(y_final, X_final).fit()
print("\nOLS summary for Vestlandet:")
print(ols_vest.summary())

# coefficients + VIF together
#coef_df = pd.DataFrame({
    #"coef": ols_vest.params,
    #"std_err": ols_vest.bse,
    #"t": ols_vest.tvalues,
    #"p": ols_vest.pvalues,
#}).join(vif_table.set_index("variable"), how="left")

#print("\nCoefficients with VIF (Vestlandet):")
#print(coef_df)
```
